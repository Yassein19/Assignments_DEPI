```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv(r'Titanic.csv')
```

```python
sex_counts = df["Sex"].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'pink'])
plt.title("Sex Distribution")
plt.show()
```

```python
# Ensure proper data types
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Fare"] = pd.to_numeric(df["Fare"], errors="coerce")
df = df.dropna(subset=["Age", "Fare"])  # Remove missing values for numerical analysis

# Set style
sns.set(style="whitegrid")

# Create a figure with multiple subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # 3x3 grid

#  1. Survival Count (Bar Chart)
df_surv = df["Survived"].value_counts()
axes[0, 0].bar(df_surv.index, df_surv.values, color=["red", "green"])
axes[0, 0].set_title("Survival Count")
axes[0, 0].set_xticks([0, 1])
axes[0, 0].set_xticklabels(["Not Survived", "Survived"])
axes[0, 0].set_ylabel("Count")

#  2. Passenger Class Distribution (Bar Chart)
df_pclass = df["Pclass"].value_counts().sort_index()
axes[0, 1].bar(df_pclass.index, df_pclass.values, color="blue")
axes[0, 1].set_title("Passenger Class Distribution")
axes[0, 1].set_xlabel("Class")
axes[0, 1].set_ylabel("Count")

#  3. Gender Distribution (Bar Chart)
df_sex = df["Sex"].value_counts()
axes[0, 2].bar(df_sex.index, df_sex.values, color=["lightblue", "pink"])
axes[0, 2].set_title("Gender Distribution")
axes[0, 2].set_ylabel("Count")

#  4. Age Distribution (Histogram)
axes[1, 0].hist(df["Age"], bins=30, color="purple", edgecolor="black", alpha=0.7)
axes[1, 0].set_title("Age Distribution")
axes[1, 0].set_xlabel("Age")
axes[1, 0].set_ylabel("Count")

#  5. Fare Distribution (Histogram)
axes[1, 1].hist(df["Fare"], bins=30, color="orange", edgecolor="black", alpha=0.7)
axes[1, 1].set_title("Fare Distribution")
axes[1, 1].set_xlabel("Fare")
axes[1, 1].set_ylabel("Count")

#  6. Embarkation Port (Bar Chart)
df_embarked = df["Embarked"].value_counts()
axes[1, 2].bar(df_embarked.index, df_embarked.values, color="green")
axes[1, 2].set_title("Embarkation Port")
axes[1, 2].set_xlabel("Port")
axes[1, 2].set_ylabel("Count")

#  7. SibSp (Siblings/Spouse Count) (Bar Chart)
df_sibsp = df["SibSp"].value_counts().sort_index()
axes[2, 0].bar(df_sibsp.index, df_sibsp.values, color="brown")
axes[2, 0].set_title("Number of Siblings/Spouses")
axes[2, 0].set_xlabel("SibSp")
axes[2, 0].set_ylabel("Count")

#  8. Parch (Parents/Children Count) (Bar Chart)
df_parch = df["Parch"].value_counts().sort_index()
axes[2, 1].bar(df_parch.index, df_parch.values, color="cyan")
axes[2, 1].set_title("Number of Parents/Children")
axes[2, 1].set_xlabel("Parch")
axes[2, 1].set_ylabel("Count")

#  9. Age vs Fare (Scatter Plot)
axes[2, 2].scatter(df["Age"], df["Fare"], color="red", alpha=0.5)
axes[2, 2].set_title("Age vs Fare")
axes[2, 2].set_xlabel("Age")
axes[2, 2].set_ylabel("Fare")

# Adjust layout
plt.tight_layout()
plt.show()
```
